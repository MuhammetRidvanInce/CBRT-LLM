```python
import requests
from bs4 import BeautifulSoup
import os


# Ana sayfa URL'si
yıl = "2014/"

base_url = "https://www.tcmb.gov.tr/wps/wcm/connect/EN/TCMB+EN/Main+Menu/Core+Functions/Monetary+Policy/Monetary+Policy+Committee/"+yıl
link_initial = "https://www.tcmb.gov.tr"
headers = {"User-Agent": "Mozilla/5.0"}


download_folder = "tcmb_ppk"
os.makedirs(download_folder, exist_ok=True)

response = requests.get(base_url, headers=headers)
soup = BeautifulSoup(response.content, "html.parser")
karar_divs = soup.find_all("div", id="y00010001")[0].find_all("div")[0].find_all("a", class_ = "type-link")
ozet_divs = soup.find_all("div", id="y00010002")[0].find_all("div")[0].find_all("a", class_ = "type-link")

for i, d in enumerate(karar_divs):

    pdf_url = link_initial + d["href"]

    name = d["title"].split()[1]
    pdf_name = f"Decision_{yıl[:-1]}_{i+1}_{name}.pdf"

    save_path = os.path.join(download_folder, pdf_name)
    pdf_response = requests.get(pdf_url, headers=headers)
    
    print(f"İndiriliyor: {pdf_name}")
    try:
        pdf_response = requests.get(pdf_url, headers=headers)
        if pdf_response.status_code == 200:
            with open(save_path, "wb") as f:
                f.write(pdf_response.content)
        else:
                print(f"Hata [{pdf_response.status_code}]: {pdf_url}")
    except Exception as e:
        print(f"İndirme hatası: {e}")

for i, d in enumerate(ozet_divs):

    pdf_url = link_initial + d["href"]

    name = d["title"].split()[1]
    pdf_name = f"Summary_{yıl[:-1]}_{i+1}_{name}.pdf"

    save_path = os.path.join(download_folder, pdf_name)
    pdf_response = requests.get(pdf_url, headers=headers)
    
    print(f"İndiriliyor: {pdf_name}")
    try:
        pdf_response = requests.get(pdf_url, headers=headers)
        if pdf_response.status_code == 200:
            with open(save_path, "wb") as f:
                f.write(pdf_response.content)
        else:
                print(f"Hata [{pdf_response.status_code}]: {pdf_url}")
    except Exception as e:
        print(f"İndirme hatası: {e}")

print("Tüm PDF dosyaları indirildi.")
```

```python
import requests
from bs4 import BeautifulSoup
import os

# Ana sayfa URL'si
yıl = "2025/"

base_url = "https://www.tcmb.gov.tr/wps/wcm/connect/EN/TCMB+EN/Main+Menu/Core+Functions/Monetary+Policy/Monetary+Policy+Committee/"+yıl
link_initial = "https://www.tcmb.gov.tr"
headers = {"User-Agent": "Mozilla/5.0"}

download_folder = "tcmb_ppk"
os.makedirs(download_folder, exist_ok=True)

response = requests.get(base_url, headers=headers)
soup = BeautifulSoup(response.content, "html.parser")

divs = soup.find_all("tbody")[0].find_all("tr")

for i, d in enumerate(divs):

    r = d.find_all("td")
    

    karar_page = r[0].find_all("a")[0]["href"]
    # print(karar_page)
    if karar_page[0] =="h":
        karar_page_url = karar_page
    else:
        karar_page_url = link_initial + karar_page
    
    response = requests.get(karar_page_url, headers=headers)
    soup = BeautifulSoup(response.content, "html.parser")
    url = soup.find_all('div', id ="tcmbMainContent")[0].find_all("div")[1].find_all("a")[0]["href"]
    karar_pdf_url = link_initial + url
    # # karar_pdf_name = f"ppk_karar_{yıl[:-1]}_{i+1}.pdf"  # ya da a_tag.get_text(strip=True).replace(" ", "_") + ".pdf"

    
    # name = d.text.split()[1]
    name = d.text.split()[0] # if year = 2025 
    karar_pdf_name = f"Decision_{yıl[:-1]}_{i+1}_{name}.pdf"
    

    karar_save_path = os.path.join(download_folder, karar_pdf_name)

    print(f"İndiriliyor: {karar_pdf_name}")
    try:
        pdf_response = requests.get(karar_pdf_url, headers=headers)
        if pdf_response.status_code == 200:
            with open(karar_save_path, "wb") as f:
                f.write(pdf_response.content)
        else:
            print(f"Hata [{pdf_response.status_code}]: {karar_pdf_url}")
    except Exception as e:
        print(f"İndirme hatası: {e}")

    

    ozet_page = r[1].find_all("a")[0]["href"]
    if ozet_page[0] =="h":
        ozet_page_url = ozet_page
    else:
        ozet_page_url = link_initial + ozet_page

    response = requests.get(ozet_page_url, headers=headers)
    soup = BeautifulSoup(response.content, "html.parser")
    url = soup.find_all('a', class_ ="pdf type-link")[0]["href"]

    ozet_pdf_url = link_initial + url

    # name = d.text.split()[1]
    name = d.text.split()[0] # if year = 2025 
    ozet_pdf_name = f"Summary_{yıl[:-1]}_{i+1}_{name}.pdf"

    ozet_save_path = os.path.join(download_folder, ozet_pdf_name)

    print(f"İndiriliyor: {ozet_pdf_name}")
    try:
        pdf_response = requests.get(ozet_pdf_url, headers=headers)
        if pdf_response.status_code == 200:
            with open(ozet_save_path, "wb") as f:
                f.write(pdf_response.content)
        else:
            print(f"Hata [{pdf_response.status_code}]: {ozet_pdf_url}")
    except Exception as e:
        print(f"İndirme hatası: {e}")
```
